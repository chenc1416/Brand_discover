```python
def brand_fshop(df):
    dfs = list(df)
    brand_fshop = []
    brand_name = re.compile(r'(.*)(官方旗舰店)')
    
    for shop_name in dfs:
        try: 
            name = re.search(brand_name, shop_name).group(1)
        except:
            pass
        if name not in brand_fshop:
            brand_fshop.append(name)
    
    
    return brand_fshop
        
```

```python
brand_fsh =  brand_fshop(df['shop_name'].fillna('NaN'))
print(brand_fsh)
```

```python
def brand_fsearch(df):
    bfs = []
    nobrand = re.compile(r'(包|香|鞋|裤|戒|表|妆|装|项链|本|淡斑|靴|隔离|系统|水|运动|红腰子|珠宝|首饰|耳钉|洗|沐浴露)')
    for tag in df:
        if not re.search(nobrand,tag):
            if tag not in bfs:
                bfs.append(tag)
                
    return bfs
    
```

```python
brand_fst = brand_fsearch(df['search_terms'])
print(brand_fst)
```

```python
def allbrand(list1,list2):
    A = ahocorasick.Automaton()
    for brand in list1:
        A.add_word(brand.lower(), (list1.index(brand), brand.lower()) )
    
    for brand in list2:
        if brand.lower() not in A:
            A.add_word(brand.lower(), (list2.index(brand), brand.lower()) )
            list1.append(brand)
            
    return list1
```

```python
takebrand=allbrand(brand_fsh,brand_fst)
print(takebrand)
```

```python
def gettrie(list1):
    A = ahocorasick.Automaton()
    
    for name in list1:
        A.add_word(name.lower(), (list1.index(name), name.lower()) )

    return A
```

```python
A = gettrie(takebrand)
```

```python
def getbrand(df1,df2,A):
    
    A.make_automaton()
    brand_list = []
    num = 0
    df1.fillna('NaN')
    while num <len(df1):
        if df1[num] == 'NaN':
            brand_list.append('Other Brand')
        else:
            if df2[num].lower() in A:
                brand_list.append(df2[num].upper())

            else:
                try:
                    item_list = []
                    for item in A.iter(df1[num].lower()):
                        item_list.append(list(item)[1][1])
                        brand_list.append(item_list[0].upper())
                except:
                    brand_list.append('Other Brand')
                    pass
        
        num += 1
            
    return brand_list
                
```

```python
 brands = getbrand(df['product_name'],df['search_terms'],A)
```
